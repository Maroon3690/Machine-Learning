### Matrix Multiplication
In Machine Learning, matrix multiplication is one of the most common tensor operation

There are two rules our tensors (or matrices) need to fulfill if we're going to matrix multiply them
1. the Inner Dimension must match
2. The resulting matrix has the shape of the outer dimension

```python
# Matrix multliplication in TensorFlow
print(tensor)
tf.matmul(tensor, tensor)
```

```python
tensor * tensor
```

```python
# Matrix Multliplication with Python Operator @
tensor @ tensor
```

```python
tensor.shape
```

```python
# Create a tensor of (3,2) tensor
X = tf.constant([[1, 2],
                 [3, 4],
                 [5, 6]])
# Create another (3,2) tensor
Y = tf.constant ([[7, 8],
                 [9, 10],
                 [11,12]])
X, Y
```

```python
# Try to matrix multply  tensors of same shape
tf.matmul(X, Y)
```

```python
# Another Method
X * Y
```

```python
# ### Reshape Y(same problem)Create a tensor of (3,2) tensor with change in shape of Y
tf.reshape(Y, shape=(2, 3))
```

```python
X.shape, tf.reshape(Y, shape=(2, 3)).shape
```

```python
# Try to  matrix multiply  X by reshaped Y
X @ tf.reshape(Y, shape=(2, 3))
```

```python
tf.matmul(X, tf.reshape(Y, shape=(2, 3)))
```

```python
### Reshape X but not Y
tf.matmul(tf.reshape(X, shape=(2, 3)), Y)
```

```python
tf.reshape(X, shape=(2, 3)).shape, Y.shape
```

```python
# Can do the same with Transpose
X, tf.transpose(X), tf.reshape(X, shape=(2, 3))
```

```python
# Try matrix multiplication with transpose rather than reshape
tf.matmul(tf.transpose(X), Y)
```

 Generally, when performing matrix multiplication on two tensors and one of the axes doesn't line up, you will transpose (rather than reshape) one of the tensors to get satisfy the matrix multiplication rules.
